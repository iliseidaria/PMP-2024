## Analiza Retelei Bayesiene pentru Clasificarea E-mailurilor

#### a) Identificarea Independentelor
* **Independente conditionate:**
  * O si L sunt independente conditionat de S. Adica, odata ce stim daca un e-mail este spam sau nu, cunoasterea faptului ca acesta contine "oferta" nu ne ofera nicio informatie suplimentara despre probabilitatea ca acesta sa contina link-uri.
* **Nici o independenta marginala:**
  * Nu exista doua noduri care sa fie complet independente una de alta, fara nicio conditionare. De exemplu, probabilitatea ca un e-mail sa aiba o lungime mare depinde atat de faptul ca este spam, cat si de prezenta link-urilor.

#### b) Clasificarea E-mailurilor
Pentru a clasifica un e-mail, trebuie sa calculam probabilitatea ca acesta sa fie spam, data fiind o anumita combinatie de valori pentru O, L si M. Aceasta se poate face folosind formula lui Bayes si propagarea probabilitatilor prin retea.

**Formula lui Bayes:**
```
P(S=1|O, L, M) = (P(O, L, M|S=1) * P(S=1)) / P(O, L, M)
```
unde:
* `P(S=1|O, L, M)` este probabilitatea ca e-mailul sa fie spam, date fiind valorile observate ale O, L si M.
* `P(O, L, M|S=1)` este probabilitatea de a observa valorile O, L si M, dat fiind faptul ca e-mailul este spam.
* `P(S=1)` este probabilitatea a priori ca un e-mail sa fie spam.
* `P(O, L, M)` este probabilitatea de a observa valorile O, L si M, indiferent daca e-mailul este spam sau nu.

**Propagarea Probabilitatilor:**
Pentru a calcula `P(O, L, M|S=1)`, putem folosi structura retelei si tabelele de probabilitati date. De exemplu, pentru o anumita combinatie de valori (O=1, L=1, M=1), am avea:
```
P(O=1, L=1, M=1|S=1) = P(O=1|S=1) * P(L=1|S=1) * P(M=1|S=1, L=1)
```
**Decizia de Clasificare:**
Dupa ce calculam `P(S=1|O, L, M)`, putem compara aceasta probabilitate cu `P(S=0|O, L, M)` (care poate fi calculata similar). Daca `P(S=1|O, L, M)` este mai mare, clasificam e-mailul ca fiind spam, altfel il clasificam ca fiind non-spam.
