Exercitiu laborator 2 - simularea variabilelor aleatoare

Urna contine:
*   3 bile rosii
*   4 bile albastre
*   2 bile negre

Reguliaruncare zar:

*   nr prim (2,3,5) -> adaugam bila neagra in urna
*   dam 6 -> adaugam o bila rosie
*   dam 1,4 -> adaugam o bila albastra



```python
#a) experiment -> returneaza bila extrasa
import random

def game():

  #starea initiala a urnei
  urna = 3* ['rosie'] + 4 * ['albastra'] + 2 * ['neagra']

  numar_zar = random.randint(1, 6)

  if numar_zar == 1:
    urna.append('albastra')
  elif numar_zar == 2:
    urna.append('neagra')
  elif numar_zar == 3:
    urna.append('neagra')
  elif numar_zar == 4:
    urna.append('albastra')
  elif numar_zar == 5:
    urna.append('neagra')
  elif numar_zar == 6:
    urna.append('rosie')

  #extrag o bila random si o returnez
  return random.choice(urna)

print("Bila extrasa este: ", game())
```

```python
#b) probabilitate frecventiala -> care este probabilitatea

# Simulare experiment de 5000 de ori
num_simulations = 5000
rezultate = []

for _ in range(num_simulations):
  culoare_bila = game()
  rezultate.append(culoare_bila)

# nr de bile rosii extrase
num_bile_rosii = rezultate.count('rosie')

# Probabilitatea de a obtine o bila rosie = de cate ori a picat rosu/nr total de extrageri de bile
probabilitate_rosie = num_bile_rosii / num_simulations

print(f"Probabilitatea de a obtine o bila rosie: {probabilitate_rosie:.4f}")
```
