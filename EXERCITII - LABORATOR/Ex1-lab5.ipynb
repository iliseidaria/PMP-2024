Se dau teste de diferite dificultati: dificil, mediu, si usor. Exista probabilitati de tranzitie intre dificultati, in functie de dificultatea testului anterior:

Daca un test este dificil, urmatorul poate fi doar mediu sau usor, fiecare cu o probabilitate de 0,5. Daca un test este mediu, urmatorul poate fi dificil cu probabilitatea 0,25, mediu cu probabilitatea 0,5, si usor cu probabilitatea 0,25. Daca un test este usor, urmatorul poate fi dificil cu probabilitatea 0,2, mediu cu probabilitatea 0,3, si usor cu probabilitatea 0,4. Notele unui student (FB, B, S, sau NS) depind de dificultatea testului:

Pentru test dificil: [0,2, 0,3, 0,4, 0,1] pentru fiecare nota.

Pentru test mediu: [0,15, 0,25, 0,5, 0,1].

Pentru test usor: [0,2, 0,3, 0,4, 0,1].

a) Folosind hmmlearn, sa definim modelul Markov ascuns si sa desenam diagrama de stare.

b) Sa calculam cea mai probabila secventa de dificultati pentru o secventa de note data: FB, FB, S, B, B, S, B, B, NS, B, S.

Raspuns:

Definirea Modelului Markov Ascuns folosind hmmlearn Pentru a defini modelul, avem nevoie de:

Matricea de tranzitie intre starile de dificultate:

Dificil -> [0, 0.5, 0.5] Mediu -> [0.25, 0.5, 0.25] Usor -> [0.2, 0.3, 0.4] Matricea de emisie, bazata pe probabilitatile de notare in functie de dificultatea testului:

Pentru Dificil: [0.2, 0.3, 0.4, 0.1] Pentru Mediu: [0.15, 0.25, 0.5, 0.1] Pentru Usor: [0.2, 0.3, 0.4, 0.1] Vectorul de probabilitati initiale (presupunem ca probabilitatile pentru primul test sunt egale):[1/3, 1/3, 1/3]

```python
import numpy as np
from hmmlearn import hmm

states = ["Dificil", "Mediu", "Usor"]
number_states = len(states)

observations = ["FB", "B", "S", "NS"]
number_observations = len(observations)

transition_matrix = [
    [0.0, 0.5, 0.5],
    [0.25, 0.5, 0.25],
    [0.2, 0.3, 0.5]
]

emission_matrix = [
    [0.2, 0.3, 0.4, 0.1],
    [0.15, 0.25, 0.5, 0.1],
    [0.2, 0.3, 0.4, 0.1]
]
start_probability = [1/3, 1/3, 1/3]

model = hmm.CategoricalHMM(n_components=number_states, init_params="")
model.startprob_ = start_probability
model.transmat_ = transition_matrix
model.emissionprob_ = emission_matrix

observations_map = {obs: idx for idx, obs in enumerate(observations)}
sequence = [observations_map[note] for note in ["FB", "FB", "S", "B", "B", "S", "B", "B", "NS", "B", "S"]]

logprob, state_sequence = model.decode(np.array(sequence).reshape(-1, 1), algorithm="viterbi")

most_probable_sequence = [states[state] for state in state_sequence]
print("Cea mai probabila secventa de dificultati este:", most_probable_sequence)
```
