### Ex. 1. (formula lui Bayes). Să presupunem că investigăm o boală (B), care afectează doar un procent din populaţie (P(B) = 0.01). Avem un test de diagnosticare pentru această boală, care are următoarele caracteristici:
● Sensibilitate (probabilitatea ca testul să fie pozitiv dacă persoana este bolnavă): P(Test = Pozitiv∣B) = 0.95
● Specificitate (probabilitatea ca testul să fie negativ dacă persoana nu este bolnavă): P(Test = Negativ∣¬B) = 0.90
a) Dacă o persoană este testată şi rezultatul este pozitiv, care este probabilitatea ca aceasta să fie efectiv bolnavă?
Cum explicaţí rezultatul?
b) Care ar trebui să fie specificitatea minimă pentru ca probabilitatea de mai sus să ajungă la 50%?
Încărcaţi argumentul fie în Markdown/Latex, fie ca foto la foia de lucru.

Pentru a rezolva acest exercițiu, aplicăm formula lui Bayes pentru a calcula probabilitatea ca o persoană să fie bolnavă, știind că rezultatul testului este pozitiv.

#### Date inițiale
- Probabilitatea de a avea boala: \( P(B) = 0.01 \)
- Sensibilitatea testului: $ P(\text{Test} = \text{Pozitiv} \mid B) = 0.95 $
- Specificitatea testului: $ P(\text{Test} = \text{Negativ} \mid \neg B) = 0.90 $

#### a) Calculul probabilității ca o persoană să fie bolnavă dacă testul este pozitiv

Aplicăm formula lui Bayes:
$
P(B \mid \text{Test} = \text{Pozitiv}) = \frac{P(\text{Test} = \text{Pozitiv} \mid B) \cdot P(B)}{P(\text{Test} = \text{Pozitiv})}
$

###Calculul probabilității $ P(\text{Test} = \text{Pozitiv}) $ este:

$$
P(\text{Test} = \text{Pozitiv}) = P(\text{Test} = \text{Pozitiv} \mid B) \cdot P(B) + P(\text{Test} = \text{Pozitiv} \mid \neg B) \cdot P(\neg B)
$$

unde $ P(\text{Test} = \text{Pozitiv} \mid \neg B) = 1 - \text{Specificitate} = 0.10 $ și $ P(\neg B) = 1 - P(B) = 0.99 $.

Substituim valorile:
- $ P(\text{Test} = \text{Pozitiv} \mid B) \cdot P(B) = 0.95 \cdot 0.01 = 0.0095  $
- $ P(\text{Test} = \text{Pozitiv} \mid \neg B) \cdot P(\neg B) = 0.10 \cdot 0.99 = 0.099 $

Rezultă:
$
P(\text{Test} = \text{Pozitiv}) = 0.0095 + 0.099 = 0.1085
$

Aplicăm formula lui Bayes:
$
P(B \mid \text{Test} = \text{Pozitiv}) = \frac{0.0095}{0.1085} \approx 0.0876 \quad \text{(sau 8.76%)}
$

**Explicația rezultatului**: Probabilitatea ca o persoană să fie bolnavă, știind că testul este pozitiv, este de 8.76%. Aceasta se datorează ratei scăzute a bolii (1%) și specificității relativ scăzute a testului (90%), care generează o proporție relativ mare de rezultate fals-pozitive.

#### b) Specificitatea minimă necesară pentru ca $ P(B \mid \text{Test} = \text{Pozitiv}) $ să fie de cel puțin 50%

Pentru a obține $ P(B \mid \text{Test} = \text{Pozitiv}) \geq 0.5 $, ajustăm formula lui Bayes și rezolvăm pentru specificitate.

#### Context și date inițiale
- Numărul de clienți pe oră urmează o distribuție Poisson cu parametru $ \lambda = 20 $ clienți/oră.
- Timpul de plasare și plată a unei comenzi urmează o distribuție exponențială cu media de 2 minute, deci rata $ \mu = \frac{1}{2} = 0.5 $.
- Timpul de preparare la stația de gătit urmează o distribuție exponențială cu media $ \alpha $ minute, unde $ \alpha $ este un parametru de optimizat.

#### 1. Modelul probabilistic

Pentru a modela acest context probabilistic:
- Numărul de clienți pe oră: $ N \sim \text{Poisson}(\lambda = 20) $
- Timpul de plasare a unei comenzi: $ T_\text{plasare} \sim \text{Exponențial}(0.5) $
- Timpul de preparare a comenzii la stația de gătit: $ T_\text{preparare} \sim \text{Exponențial}\left(\frac{1}{\alpha}\right)$

#### 2. Determinarea valorii maxime a lui \$alpha$

Se cere să determinăm valoarea maximă a lui $ \alpha $ pentru a ne asigura că toți clienții primesc mâncarea în mai puțin de 15 minute, cu o probabilitate de cel puțin 95%. Astfel, probabilitatea de a servi un client în mai puțin de 15 minute este:
$$
P(T_\text{plasare} + T_\text{preparare} < 15) \geq 0.95
$$
Aceasta se poate determina prin simulare numerică, iterând valori pentru $\alpha $ și calculând probabilitatea ca timpul total să fie sub 15 minute pentru toți clienții.

#### 3. Determinarea timpului mediu de așteptare pentru un client

Cu valoarea optimizată a lui $ \alpha $ din punctul anterior, timpul mediu total de așteptare este dat de suma mediei timpului de plasare și a mediei timpului de gătit.
